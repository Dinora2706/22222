```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

data = {"15-18":200, "19-24":80, "25 or above":50, "10-14":50}
courses = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (5, 5))


plt.bar(courses, values, color = "blue", width = 0.4)
plt.xlabel("market segment", fontsize =18, color = "black")
plt.ylabel("Expected number of customers", fontsize = 16, color = "black")
plt.title("Ð¡ustomer potential by market segment", color = "black")
plt.xticks(rotation = 90)
plt.ticklabel_format(style = "plain")

plt.show()
```
